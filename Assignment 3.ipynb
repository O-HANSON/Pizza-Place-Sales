### Joining csv files into one dataframe

```python
# Joining order_details with pizzas on pizza_id

merged = pd.merge(order_details, pizzas, on='pizza_id', how='left')

merged
```

```python
# Joining with orders on order_id

merged = pd.merge(merged, orders, on='order_id', how='left')

merged
```

```python
# Joining  with pizza_types on pizza_type_id

merged = pd.merge(merged, pizza_types, on='pizza_type_id', how='left')

merged
```

```python
# Creating a revenue column for sales analysis

merged['revenue'] = merged['quantity'] * merged['price']

merged
```

```python
# Converting  date and time to proper datetime formats for time-based analysis

merged['date'] = pd.to_datetime(merged['date'])
merged['time'] = pd.to_datetime(merged['time'], format='%H:%M:%S')

merged
```

## Analysis on Merged Dataframe

```python
# Finding total revenue/sales

total_revenue = merged['revenue'].sum()
print(f"Total Revenue: ${total_revenue:,.2f}")
```

```python
# Finding the total quantity sold

total_quantity = merged['quantity'].sum()
print(f"Total Quantity Sold: {total_quantity:,}")
```

```python
# Finding the total orders

total_orders = orders['order_id'].nunique()
print(f"Total Orders: {total_orders:,}")
```

```python
# Types of pizza being sold

num_pizza_types = pizza_types['pizza_type_id'].nunique()
print(f" Number of Pizza Types: {num_pizza_types}")
```

```python
# Finding average price of pizzas

average_price = pizzas['price'].mean()
print(f"Average Pizza Price: ${average_price:.2f}")
```

```python
# Finding sales peak hours

# Quick & safe version - use this
orders['time'] = orders['time'].astype(str).str.strip()
orders['hour'] = orders['time'].str[:2].astype(int)
peak_hours = orders.groupby('hour').size().sort_values(ascending=False).head(5)

print("sales peak hours(Top 5):")
print(peak_hours)
```

```python
# Finding total sales on each day of the week

merged['day_of_week'] = merged['date'].dt.day_name()
sales_by_day = merged.groupby('day_of_week')['revenue'].sum().sort_values(ascending=False)
print("Total Sales on each Day of the Week:")
print(sales_by_day)
```

```python
# Day of the Week with most sales

most_sales_by_day = sales_by_day.idxmax()
print(f" Day with the most sales: {most_sales_by_day} (${sales_by_day.max():,.2f})")
```

```python
# Finding top 5 best selling pizzas

top_pizzas_rev =(merged.groupby('name')['revenue'].sum().sort_values(ascending=False).head(5))

print("Top 5 Best Selling Pizzas ")
top_pizzas_rev
```

```python
# Finding Sales Made In Each Month

merged['month'] = merged['date'].dt.to_period('M')
sales_by_month = merged.groupby('month')['revenue'].sum().sort_index()
print("Sales In Each Month")
sales_by_month
```

```python
# Plotting to see noticeable trend

sales_by_month.plot(kind='line', title='Monthly Sales Trend')

plt.show()
```

#### Noticeable trend is July recorded the highest sales ( 72557.90 ) , followed by March ( 70397.10 ) followed by November ( 70395.35 ) and it follows in that order.

```python
# Finding underperforming pizza types

underperforming = (merged.groupby('name')['quantity'].sum().sort_values(ascending=True).head(10))
print("Underperforming Pizza Types (Lowest Quantity Sold - Top 10)")
underperforming
```

```python
# Plotting to see trend in top 10 underperforming pizza types

underperforming.plot(kind='line', title='Top 10 Underperforming Pizza Types')
plt.xticks(rotation=90)

plt.show()
```

#### Additional data exploration

```python
# Finding additonal revenue per order, by the day of the week

average_rev_per_day = merged.groupby('day_of_week')['revenue'].sum() 
average_rev_per_day = average_rev_per_day.sort_values(ascending=False)

print("Average Revenue Per Day of the week")
print(average_rev_per_day.round(2))
print("Highest Average Revenue per day", average_rev_per_day.idxmax(), f"${average_rev_per_day.max():.2f}")
```

```python
# Finding most popular pizza size sold , by quantity

size_popularity = merged.groupby('size')['quantity'].sum().sort_values(ascending=False)

print("Total quantity sold by pizza size:")
size_popularity
```

```python
print("Most popular size", size_popularity.idxmax(), f"({size_popularity.max():,} pizzas)")
```

```python
size_popularity.plot(kind='line', title='Total Quantity Sold by Pizza Size')

plt.show()
```

```python
# Finding Orders that include multiple different sizes in the same order

sizes_per_order = merged.groupby('order_id')['size'].nunique()
multiple_size_orders = sizes_per_order[sizes_per_order > 1]

print(f"Orders with more than 1 different size: {len(multiple_size_orders):,}")
print(f"Percentage of orders with multiple sizes: {(len(multiple_size_orders) / len(sizes_per_order) * 100):.1f}%")
```

```python
print("How many sizes per order ")
print(sizes_per_order.value_counts().head(6))
```

```python
# Plotting Orders with multiple different sizes

size_counts = sizes_per_order.value_counts()  

plt.pie(size_counts, labels=size_counts.index.astype(str),autopct='%1.1f%%', colors=["navy", "lightblue", "lightgreen", "salmon", "plum"])
plt.title("Orders by Number of Different Pizza Sizes")
plt.show()
```
